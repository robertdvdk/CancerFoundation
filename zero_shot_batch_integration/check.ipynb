```python
import pandas as pd
df1 = pd.read_csv('./data/neftel_ss2/batch_integration_evaluation.csv', index_col=0)
df1 = df1.rename(columns={'Unnamed: 1': 'statistic'})
df1[df1['statistic'].isin(['mean', 'std'])]
```

```python
import pandas as pd
df2 = pd.read_csv('./data/kim_lung/batch_integration_evaluation.csv', index_col=0)
df2 = df2.rename(columns={'Unnamed: 1': 'statistic'})
df2[df2['statistic'].isin(['mean', 'std'])]
```

```python
import pandas as pd
df3 = pd.read_csv('./data/ji_skin/batch_integration_evaluation.csv', index_col=0)
df3 = df3.rename(columns={'Unnamed: 1': 'statistic'})
df3[df3['statistic'].isin(['mean', 'std'])]
```

```python
print(df3)
```

```python
import matplotlib.pyplot as plt
# Plotting
plt.figure(figsize=(10, 6))

dataframes = {'Neftel Brain': df1, 'Kim Lung': df2, 'Ji Skin': df3}
colors = {'Neftel Brain': 'blue', 'Kim Lung': 'green', 'Ji Skin': 'red'}

for name, df in dataframes.items():
    # Reset index to work with columns
    df_reset = df.reset_index()
    
    # Extract epoch number
    df_reset['epoch'] = df_reset['Embedding'].str.split('_').str[1].astype(int)
    
    # Get mean and std for the 'Total' score
    total_mean = df_reset[df_reset['statistic'] == 'mean'].set_index('epoch')['Total']
    total_std = df_reset[df_reset['statistic'] == 'std'].set_index('epoch')['Total']
    
    # Sort by epoch for correct plotting
    total_mean = total_mean.sort_index()
    total_std = total_std.sort_index()
    
    epochs = total_mean.index
    
    # Plot mean line
    plt.plot(epochs, total_mean, label=f'{name} Mean Total Score', color=colors[name])
    
    # Plot standard deviation as a shaded area
    plt.fill_between(epochs, total_mean - total_std, total_mean + total_std, color=colors[name], alpha=0.2, label=f'{name} Std Dev')

# Add titles and labels
plt.title('Batch integration score vs. Epochs for 3 cancer datasets')
plt.xlabel('Epochs')
plt.ylabel('Batch integration score')
plt.legend()
plt.grid(True)

# Save the plot
plt.savefig('total_score_vs_epochs.png')

print("Plot saved as total_score_vs_epochs.png")
```
